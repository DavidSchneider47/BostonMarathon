```python
#scale the data
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
scaled_array = min_max_scaler.fit_transform(df2)
```

```python
#view the new scaled data set 
df_normalized = pd.DataFrame(scaled_array)
df_normalized
```
